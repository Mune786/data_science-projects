# scatter plot bw age and fare
## A **scatter plot** is used to visualize the relationship between two continuous variables. It’s applied to identify trends, correlations, clusters, or outliers in the data.

# feature enginring
##add new cols in data
### **Feature engineering** is the process of creating, selecting, or transforming variables (features) to improve the performance of machine learning models. It’s applied to enhance model accuracy, handle missing data, encode categorical variables, or extract meaningful patterns from raw data.

```python


# Handle missing values in 'age' before binning
df['age'].fillna(value=df['age'].mean(), inplace=True)

# Define bins and labels
bins = [0, 1, 5, 12, 18, 30, 50, 80]
labels = ['infant', 'toddlers', 'kids', 'teens', 'young', 'm_age', 'old']

# Create a new column with binned age
df['binned_age'] = pd.cut(df['age'], bins=bins, labels=labels)

# Display the DataFrame with the new column
print(df[['age', 'binned_age']].head())
```

```python
# Count the occurrences of each bin
bin_counts = df['binned_age'].value_counts()

# Display the counts
print(bin_counts)
```

```python


# Ensure 'age_groups' column exists (if it's a binned column, create it first)
bins = [0, 1, 5, 12, 18, 30, 50, 80]
labels = ['infant', 'toddlers', 'kids', 'teens', 'young', 'm_age', 'old']
df['age_groups'] = pd.cut(df['age'].fillna(df['age'].mean()), bins=bins, labels=labels)

# Select specific columns
df_01 = df[['survived', 'age_groups', 'fare', 'class']]

# Display the new DataFrame
print(df_01.head())
```
